```python
import pandas as pd
from sklearn.linear_model import LinearRegression
df=pd.read_csv("house.csv")
print(df)
df.head()
df.info()
```

```python
X = df[["area"]]
y = df["price"]
model = LinearRegression()
model.fit(X, y)
model.predict(pd.DataFrame({"area": [900]}))
```
